```python
import pandas as pd
```

```python
df=pd.read_csv("/content/ch.csv")
display(df.head(10))
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['day'],df['tip'])
plt.title("Scatter plot")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.colorbar()
plt.show()
```

```python
plt.plot(df['tip'])
plt.plot(df['size'])
plt.title("Line plot")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.show()
```

```python
plt.bar(df['day'],df['tip'])
plt.title("Bar chart")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.show()
```

```python
plt.hist(df['tip'])
plt.title("Histogram of tips")
plt.xlabel("tip")
plt.ylabel("Frequency")
plt.show()
```
