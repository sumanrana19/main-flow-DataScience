```python
import pandas as pd
```

```python
try:
    df = pd.read_csv('/content/USvideos.csv', encoding='ISO-8859-1', engine='python', on_bad_lines='skip')
except pd.errors.ParserError:
    print("There was an error parsing the CSV file.")
```

```python
#checking missing values
missing_values = df.isnull().sum()
print("missing values:\n", missing_values)
```

```python
#summary statistics
summary_statistics = df.describe(include='all')
print("\nsummary statistics:\n", summary_statistics)
```

```python
#styles for plot and plot distribution of numeric values
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
```

```python
#views
sns.histplot(df['views'], bins=50, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Views')
```

```python
#likes
sns.histplot(df['likes'], bins=50, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Likes')
```

```python
#dislikes
sns.histplot(df['dislikes'], bins=50, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Dislikes')
```

```python
#comment count
sns.histplot(df['comment_count'], bins=50, kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Comment Count')
```

```python
plt.tight_layout()
plt.show()
```

```python
#Identify outliers using boxplots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

#views
sns.boxplot(x=df['views'], ax=axes[0, 0])
axes[0, 0].set_title('boxplot of views')

# ikes
sns.boxplot(x=df['likes'], ax=axes[0, 1])
axes[0, 1].set_title('boxplot of likes')

#dislikes
sns.boxplot(x=df['dislikes'], ax=axes[1, 0])
axes[1, 0].set_title('boxplot of dislikes')

#comment count
sns.boxplot(x=df['comment_count'], ax=axes[1, 1])
axes[1, 1].set_title('boxplot of comment count')

plt.tight_layout()
plt.show()
```

```python
#check for correlations between numerical variables
correlation_matrix = df[['views', 'likes', 'dislikes', 'comment_count']].corr()
```

```python
#plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('correlation matrix')
plt.show()
```
