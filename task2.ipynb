Loading CSV file into pandas DataFrame

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/cvFile.csv')
df.head()
```

```python
df.columns = df.columns.str.strip()
```

Handling Missing Values

```python
df['UCZAA'].fillna(df['UCZAA'].mean(), inplace=True)
df['SulphidityL-4'].fillna(df['SulphidityL-4'].mean(), inplace=True)
```

filtered DataFrame where Y-Kappa is greater than 25 and summary statistics

```python
filtered_df = df[df['Y-Kappa'] > 25]
summary_statistics = df.describe()
```
